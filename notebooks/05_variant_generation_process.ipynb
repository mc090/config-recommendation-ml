# Variant Generation Process

This notebook reproduces the CLI-driven generation of the 5 dataset variants:
`original`, `corr_070`, `corr_060`, `dist_expert`, and `dom_<threshold>`.

```python
from pathlib import Path

import pandas as pd

from src.experiments.build_variants import build_variants
```

```python
dataset_output_dir = Path("data/processed/nb_test")

build_variants(dataset_output_dir)
```

```python
version_dirs = sorted(path for path in dataset_output_dir.glob("v*") if path.is_dir())
latest_version = version_dirs[-1]
variants_dir = latest_version / "variants"

print(f"Latest dataset version: {latest_version.name}")
print(f"Variants directory: {variants_dir}")
for dataset_path in sorted(variants_dir.glob("*/dataset.csv")):
    print(" -", dataset_path.relative_to(dataset_output_dir))
```
