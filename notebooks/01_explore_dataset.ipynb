# Dataset exploration
This notebook runs the project pipeline steps in order:
1. fetch raw data
2. extract structure
3. compute features
4. build dataset. 

It shows mocked process of dataset extraction.

```python
from src.config import Settings
from src.data.fetch_raw import fetch_raw

nb_cfg = Settings(
    max_repos=5,
    languages=["Python"],
    min_stars=500,
)

fetch_raw(cfg=nb_cfg, output_path="data/raw/nb_test.json")
```

```python
import json
from pathlib import Path

raw_file = Path("data/raw/nb_test.json")

with open(raw_file) as f:
    data = json.load(f)

if isinstance(data, list) and len(data) > 0:
    first_entry = data[1]
elif isinstance(data, dict):
    first_entry = next(iter(data.values())) if data else None
else:
    first_entry = None

if first_entry:
    print(json.dumps(first_entry, indent=2))
else:
    print("No data found in nb_test.json")
```
